```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from math import cos
import numpy as np
```

<strong> i. Define surface parametrization </strong> </br>
Find x,y,z coordinates with radius 80 from centre (x0,z0,y0 = 0)

```python
#set radius 
r = 80

#use np.linspace to define x and y coordinates
x_values = np.linspace(-r,r,1000)
y_values = np.linspace(-r,r,1000)
x,y = np.meshgrid(x_values, y_values)

#find z coordinate as a function of x and y coordinates
#x^2+y^2+z^2 = 80^2 and we want the positive hemisphere,therefore z = sqrt(80-x^2-y^2)
z = np.sqrt(80**2-y**2-x**2)
```

<strong> ii. Find intensity at each point </strong> </br>
Using dot-product of the normal of the surface with the lightning direction. </br>
Illumination direction I = (0.12,0,0.96)

```python
Illum = [0.12,0,0.96]

#function to find the normal vector
def findNormal(x,y):
    normal_x = -x/(np.sqrt(80**2-x**2-y**2))
    normal_y = -y/(np.sqrt(80**2-x**2-y**2))
    return ([normal_x,normal_y,1])

#function to normalise a vector
def normaliseVector(A):
    return A/np.linalg.norm(A)

#function to find illumination value from direction A at point B
def Illumination(A,B):
    mag_A = np.sqrt(A[0]**2+A[1]**2+A[2]**2)
    mag_B = np.sqrt(B[0]**2+B[1]**2+B[2]**2)
    normalise_A = normaliseVector(A)
    normalise_B = normaliseVector(B)
    cosine_theta = np.dot(normalise_A,normalise_B)
    return mag_A*mag_B*cosine_theta
```

```python
#create array of intensities per each coordinate, this will be called color_dimension.

color_dimension = []

for i in range(len(x)):
    color_dimension2 = []
    for j in range(len(x[0])):
        I = Illumination(Illum,[x[i][j],y[i][j],z[i][j]])
        #where z value is nan, we will put 0 as a placeholder noting that this will be inside the sphere 
        #and not in the surface i.e. not seen anyway
        if np.isnan(I):
            I = 0
        color_dimension2.append(I)
    color_dimension.append(color_dimension2)
    

#plot sphere using color_dimension as an input to the color intensity
print("Figure 1",": Illumination direction ", Illum)
plt.imshow(np.real(color_dimension), 'gray')
```
